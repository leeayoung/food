range() 함수와 유사한 frange() 함수를 다음 조건을 참고하여 만드시오.

- 1) frange()의 인자 구성은 다음과 같이 range와 동일하지만 각 인수들은 음수를 받지 않는다고 가정한다.
        range(stop)
        range(start, stop[, step])
- 2) frange() 함수의 인자에 대한 기본 시작(start) 값은 0.0이고, 기본 단계(step) 값은 0.1이다.
- 3) frange 사용 예
    - 3-1) frange(0.5)
        [0.0, 0.1, 0.2, 0.3, 0.4]
    - 3-2) frange(1.0, 2.0)
        [1.0, 1.1, 1.2, 1.3, 1.4, 1.5, 1.6, 1.7, 1.8, 1.9]
    - 3-3) frange(2.2, 4.0, 0.5)
         [2.2, 2.7, 3.2, 3.7]

```python
import math

def frange(start = 0.0, stop = None, step = 0.1):
    f_list = []
    if(stop == None) :
        stop = start
        start = 0.0
        while (start < stop):
            f_list.append(round(start,1))
            start += step
    else:
        while (start < stop):
            f_list.append(round(start,1))
            start += step
            
    return f_list
            
print frange(0.5)
print frange(1.0, 2.0)
print frange(2.2, 4.0, 0.5)
```

- frange 함수를 선언해준다. 이때 start,stop,step 값을 갖게된다.
- frange 함수의 결과 리스트를 가질 f_list를 선언한다.
- 만약 stop 값이 none이 들어온다면 원래 입력값이 stop 값이고, start 값이 0.0이 되는 것이므로 stop 값에 start 값을 넣어주고, start 값을 0.0으로 입력한다.
- 리스트를 step값에 맞춰 숫자를 추가한다.
- 이때 round함수를 이용하여 반올림하여 소수를 출력한다.

    - sum()
            0

            3
    - sum(1, 2, 3, 4, 5)
            15
    - sum(1, 5, 7, 2, -10)

```python
def sum(*num):
    sum_=0
    for x in num:
        sum_ += x
    return sum_
print sum()
print sum(1, 2)
print sum(1, 2, 3, 4, 5)
print sum(1,5,7,2,-10)
```

2) myinitial() 함수 사용 예
      - myinitial("as soon as possible")
               - ASAP

```python
def myinitial(string):
    join_li=[]
    word = string.split()
    for x in word:
        Upper = map(lambda a: a.upper(),x)
        join_li.append(Upper[0])
    print "".join(join_li)

myinitial("as soon as possible")
```

- string 값을 받아와 첫단어를 대문자로 만들어준다.

### 6번
음이 아닌 정수 n를 입력받으면 n! (factorial)을 계산하는 myfact() 함수를 재귀적 함수로 구현하시오.

- [옵션] 참을 수 있는 정도 만큼의 수행시간을 직접 기다려보면서 n을 늘려보도록 합시다.

```python
def myfact(n):
    if(n==0):
        return 1
    if(n==1):
        return 1
    else:
        return n*myfact(n-1)
    
print myfact(976)
print myfact(977)
```

- 재귀적 방법을 이용하여 팩토리얼 계산을 하도록 해주었다.
- 976까지는 동작되나 977은 오류가 발생한다.

- 3) 단어의 출현빈도가 담긴 사전 객체를 위 HTML 소스 파일과 동일한 폴더에 파일로 저장하시오.

    - URL이 http://URL 이라면 사전 객체를 담고 있는 파일명은 URL.words.frequency 이다.
    - 예: URL이 http://www.cnn.com 이라면 파일명은 www.cnn.com.words.frequency 이다.
- 4) 최소 5개 이상의 영어 웹 사이트에 대한 HTML 소스 파일과 단어 출현빈도 파일을 저장하시오.
    - 즉, 총 10개의 파일을 동일한 폴더에 생성하시오.
    - [주의] 영어 웹사이트 URL로만 5개 이상
- 5) 위 문제에서 저장한 모든 words.frequency 파일들을 객체로 다시 로드하여 본인이 저장하여 분석한 사이트들에서 가장 많이 출현한 단어 3개를 뽑아 제시하시오.
    - 반드시 pickle 모듈로 저장한 5개 이상의 words.frequency를 다시 5개 이상의 사전 객체로 로드 하는 코드가 추가되어야 함

```python
import urllib2
import string
import pickle



def tag_D(string):
    for i in range(0, string.count('<!--')):
        if string.find("<!--") != -1:
            string = string.replace(string[string.find("<!--"):string.find("-->")+3:]," ")
    for i in range(0,string.count("<")):
        if string.find("<") != -1:
            string = string.replace(string[string.find("<"):string.find(">")+1:]," ")
    
    return string


def java_D(string):
    for i in range(0, string.count('<script')):
        if string.find("<script") != -1:
            string = string.replace(string[string.find("<script"):string.find("</script>")+9:]," ")
    return string


def style_D(string):
    for i in range(0, string.count('<style')):
        if string.find("<style") != -1:
            string = string.replace(string[string.find("<style"):string.find("</style>")+8:]," ")
    return string


def punc_D(source):
    punctuation = string.punctuation
    for i in range (len(punctuation)):
        source= source.replace(punctuation[i]," ")
        
    return source


def stopword_D(words):
    stopword = [ 'A','a', 'and', 'are', 'as', 'at', 'be', 'but','But', 'by', 'for', 'if','If', 'in', 'into', 'is', 'it','It', 'no','No', 'not','Not', 'of', 'on', 'or', 's', 'such', 't', 'that','That', 'the','The', 'their','Their', 'then','Then', 'there','There', 'these','These', 'they','They', 'this','This', 'to', 'was', 'will', 'with']
    for i in range (len(stopword)):
        for j in range(len(words)):
            if words[j] == stopword[i]:
                words[j] = " "


def count_w (words):
    w_dict = {}
    for i in words:
        if w_dict.has_key(i) is False:
            w_dict[i]=1
        else:
            w_dict[i]+=1
    return w_dict
```

```python
f = open('www.nytimes.com.html','w')
f.write(urllib2.urlopen("http://www.nytimes.com").read()) 
f.close()

f = file('www.nytimes.com.html') 
source = f.read()
f.close()

source = style_D(source)
source = java_D(source)
source = tag_D(source)
source = punc_D(source)
words = source.split()  
stopword_D(words)
source = " ".join(words)
words = source.split()
words = count_w(words)

f = open('www.nytimes.com.words.frequency.txt', 'w')
pickle.dump(words, f)
f.close()
```
