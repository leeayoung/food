    - count: Counter 인스턴스가 지니는 초기 정수 값

        >>> c()

        >>> c = c + 5 
        >>> d = d - 5 
        >>> print c 
        [Count (step: 1)] 17 
        >>> print d 
        [Count (step: 2)] 9 
- 요구사항 6. 다음과 같은 관계연산 (>, <, ==)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.

        >>> print c > 10 
            True 
        >>> print d > 10 
            False 
        >>> print c < 10 
            False 
        >>> print d < 10 
            True 
        >>> print c == 17 
            True 
        >>> print d != 9 
            False 

```python
class Counter():
    # 요구사항 1
    def __init__(self, count, step = 1):
        self.count = count
        self.step = step
    # 요구사항 2
    def __str__(self):
        return '[Count (step : ' + str(self.step) + ')] ' + str(self.count)
    
    # 요구사항 3
    def incr(self):
        self.count += self.step
    
    # 요구사항 4
    def __call__(self):
        self.incr()
    
    # 요구사항 5
    def __add__(self, other):
        self.count += other
        return self
    def __sub__(self, other):
        self.count -= other
        return self
    
    # 요구사항 6
    def __lt__(self, x): 
        return self.count < x
    def __gt__(self, x): 
        return self.count > x
    def __eq__(self, x): 
        return self.count == x
    def __ne__(self, x): 
        return self.count != x    
    
        
    
#1
c = Counter(10) 
d = Counter(10, 2)
print 

#2
print c
print d
print

#3
c.incr()
d.incr()
print c
print d
print

#4
c()
d()
print c
print d
print

#5
c = c + 5 
d = d - 5
print c
print d
print

#6
print c > 10
print d > 10
print c < 10
print d < 10
print c == 17
print d != 9
```

#### 4번. 다음은 내장 자료형 list를 서브클래싱하여 만든 MySet 클래스 정의 내용이다. 다음 클래스 정의에서 __init__(), __str()__(), elimicate_duplicate()의 세 개의 메소드 코드 내용을 자신이 다른 사람에게 가르친다고 생각하며 설명해보시오.

- MySet은 집합(Set) 자료형을 정의하려는 의도하에 만들어진 클래스이다.

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print s
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print t
```

MySet은 집합(Set) 자료형을 정의하는 클래스 입니다.
- __init__()함수는 생성자 함수입니다. 생성할 때 인자로 받은 리스트의 원소들을 원소로 가지는 리스트를 만들고, MySet.eliminate_duplicate(self)를 호출합니다
- __str__() 메소드는 출력 할때 이용되는 함수입니다. Myset: { 을 문자열로 만듭니다. 다음으로 원소들을 문자열로 바꾸어 반환해줄 문자열에 추가시켜줍니다. 마지막에는 ,가 필요 없으므로 슬라이싱을 이용하여 문자열의 마지막 원소와 } 를 추가시켜 준 것을 반환합니다.
- eliminate_duplicate()함수는 중복되는 원소들을 제거해주는 함수입니다. 먼저 빈 리스트를 만들고, 인자로 들어온 리스트의 원소가 처음 만든 리스트에 없으면 추가합니다. 그리고 인자로 들어온 리스트를 비웁니다. 다시 인자로 들어왔던 리스트에 만들어 처음 만든 리스트의 원소를 차례로 집어넣어주게 되면 중복된 원소가 제거됩니다.

#### 5번. 4번 문제에 정의된 MySet 클래스에 메소드를 추가하여 다음 각 요구사항 모두를 만족시키는 코딩을 제시하시오.
- 정답을 각 요구사항별로 입력할 필요 없이 요구사항 3개 전체에 대해 1개의 MySet 클래스 정의 코드를 제시하고 주석등으로 해당 코드를 설명한다.
- 요구사항 1. | 연산으로 두 집합의 합집합을 반환한다.

        >>> u = s | t 
        >>> print u 
        MySet: {1, 2, 3, 4, 5, 6, 7, 8, 9} 
- 요구사항 2. & 연산으로 두 집합의 교집합을 반환한다.

        >>> u = s & t 
        >>> print u 
        MySet: {2, 3} 
- 요구사항 3. - 연산으로 두 집합의 차집합을 반환한다.

        >>> s = MySet([1, 2, 3]) 
        >>> t = MySet([3, 4, 5]) 
        >>> u = s - t 
        >>> print u 
        MySet: {1, 2} 

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
            
            
    # 요구사항 1
    def __or__(self,other):
        L=[]
        for x in self:
                L.append(x)
        for x in other:
                L.append(x)
        L.sort()
        return MySet(L)
    
    # 요구사항 2
    def __and__(self,other):
        L=[]
        for x in self:
            if x in other:
                if x not in L:
                    L.append(x)
        L.sort()
        return MySet(L)
    
    # 요구사항 3
    def __sub__(self,other):
        L=[]
        for x in self:
            if x not in other:
                if x not in L:
                    L.append(x)
        L.sort()
        return MySet(L)
    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print s
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print t

#1
u = s | t
print u
#2
u = s & t
print u
#3
s = MySet([1, 2, 3])
t = MySet([3, 4, 5])
u = s - t
print u
```

```python
import urllib2
import string


def tag_D(string):
    for i in range(0, string.count('<!--')):
        if string.find("<!--") != -1:
            string = string.replace(string[string.find("<!--"):string.find("-->")+3:]," ")
    for i in range(0,string.count("<")):
        if string.find("<") != -1:
            string = string.replace(string[string.find("<"):string.find(">")+1:]," ")
    
    return string


def java_D(string):
    for i in range(0, string.count('<script')):
        if string.find("<script") != -1:
            string = string.replace(string[string.find("<script"):string.find("</script>")+9:]," ")
    return string


def style_D(string):
    for i in range(0, string.count('<style')):
        if string.find("<style") != -1:
            string = string.replace(string[string.find("<style"):string.find("</style>")+8:]," ")
    return string


def punc_D(source):
    punctuation = string.punctuation
    for i in range (len(punctuation)):
        source= source.replace(punctuation[i]," ")
        
    return source


def stopword_D(words):
    stopword = [ 'A','a', 'and', 'are', 'as', 'at', 'be', 'but','But', 'by', 'for', 'if','If', 'in', 'into', 'is', 'it','It', 'no','No', 'not','Not', 'of', 'on', 'or', 's', 'such', 't', 'that','That', 'the','The', 'their','Their', 'then','Then', 'there','There', 'these','These', 'they','They', 'this','This', 'to', 'was', 'will', 'with']
    for i in range (len(stopword)):
        for j in range(len(words)):
            if words[j] == stopword[i]:
                words[j] = " "


def count_w (words):
    w_dict = {}
    for i in words:
        if w_dict.has_key(i) is False:
            w_dict[i]=1
        else:
            w_dict[i]+=1
    return w_dict

def Dic(source):
    source = style_D(source)
    source = java_D(source)
    source = tag_D(source)
    source = punc_D(source)
    words = source.split()  
    stopword_D(words)
    source = " ".join(words)
    words = source.split()
    words = count_w(words)
    return words

# 4번과제내용

class WebWordsFrequency():
    def __init__(self, *arg):
        self.URL = []
        for url in arg:
            self.URL.append(url)
    def addUrl(self, other):
        self.URL.append(other)
    def removeUrl(self, other):
        if other in self.URL:
            self.URL.remove(other)
    def listUrls(self):
        for other in self.URL:
            print other
    def getWordsFrequency(self):
        word_dic = {}
        L = []
        count = 0
        for url in self.URL:
            L = Dic(urllib2.urlopen(url).read())
            count=1
            if(count==0):
                for x in L:
                    word_dic[x] = L[x]
            else:
                for x in L:
                    if word_dic.has_key(x):
                        word_dic[x] += L[x]
                    else:
                        word_dic[x] = L[x]
        return word_dic
    def getMaxFrequencyWords(self):
        word_dic = self.getWordsFrequency()
        g = word_dic.items()
        g.sort(key=lambda x:x[1],reverse=True)
        L_word = []
        for x in range(len(g)-1):
            L_word.append(g[x])
            if(g[x][1]!=g[x+1][1]):
                break
        if(len(L_word)==0):
            return None
        print L_word
# 요구사항 2
w1 = WebWordsFrequency('http://www.espn.com/', 'http://www.bestbuy.com/', 'https://www.nasa.gov')
w2 = WebWordsFrequency('http://www.cnn.com', 'http://www.times.com')
w3 = WebWordsFrequency()
# 요구사항 3
w1.addUrl('https://github.com')
w3.addUrl('http://stackoverflow.com')
# 요구사항 4
w1.removeUrl('http://www.bestbuy.com/')
w2.removeUrl('http://stackoverflow.com')
# 요구사항 5
w1.listUrls()
# 요구사항 6
print w1.getWordsFrequency() 
# 요구사항 7
w1.getMaxFrequencyWords()
w2.getMaxFrequencyWords()
```

#### 11번.
- 위 10번 문제에서 정의한 WebWordsFrequency 클래스를 상속하여OrderedWebWordsFrequency 클래스를 정의하고 슈퍼클래스에 정의된 getWordsFrequency() 메소드를 오버라이드 하여 단어 출현 빈도를 내림 차순으로 정렬하여 리스트로 출력하시오.

    - 리스트 내의 각 원소는 단어와 빈도를 쌍으로 지니는 튜플이다.
    - getWordFrequence() 메소드에 reverse 라는 인자를 만들고 true 또는 false를 인자로 받을 수 있도록 한다.

        - reverse 인자의 디폴트 값은 false 이며, 기본적으로 내림차순으로 정렬한다.
        - reverse 인자에 true를 넣으면 오름차순으로 정렬한다.
                >>> w4 = OrderedWebWordsFrequency('http://www.times.com', 'https://www.amazon.com', 'https://github.com')

                >>> w4.getWordsFrequency()
                [('site', 12), ('science', 11), ('hello', 8), ('world', 2), ('program', 1), ('python', 1)]

                >>> w4.getWordsFrequency(reverse=true)
                [('program', 1), ('python', 1), ('world', 2), ('hello', 8), ('science', 11), ('site', 12)]

```python
class OrderedWebWordsFrequency(WebWordsFrequency):
    def getWordsFrequency(self, reverse=False):
        word_dic=WebWordsFrequency.getWordsFrequency(self)
        g = word_dic.items()
        if(reverse==False):
            g.sort(key=lambda x:x[1],reverse=True)
        else:
            g.sort(key=lambda x:x[1],reverse=False)
        return g
w4 = OrderedWebWordsFrequency('http://www.espn.com/', 'http://www.bestbuy.com/', 'https://www.nasa.gov')
print w4.getWordsFrequency()
print
print w4.getWordsFrequency(reverse=True)
```
